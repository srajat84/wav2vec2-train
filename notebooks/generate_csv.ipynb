```python
import pandas as pd
```

```python
segments = []
with open('./train_NPTEL_IITM/segments.txt') as file_local:
    segments = file_local.readlines()
```

```python
segments = [seg.split(' ') for seg in segments]
```

```python
segments
```

```python
df_segment = pd.DataFrame(segments, columns = ['segment_id', 'file_id', 'start_time', 'end_time'])
```

```python
df_segment['end_time'] = df_segment['end_time'].str.strip()
```

```python
df_segment.head()
```

```python
text = []
with open('./train_NPTEL_IITM/text.txt') as file_local:
    text = file_local.readlines()
```

```python
text[-5:]
```

```python
text[:5]
```

```python
text = [local_text.split('\t',1) for local_text in text]
```

```python
df_text = pd.DataFrame(text, columns = ['segment_id', 'transcription'])
df_text['transcription'] = df_text['transcription'].str.strip() 
df_text.head()
```

```python
df_text.tail()
```

```python
df_text.head().transcription.values
```

```python
utt2spk = []
with open('./train_NPTEL_IITM/utt2spk.txt') as file:
    utt2spk = file.readlines()
```

```python
utt2spk
```

```python
utt2spk = [utt.split(' ') for utt in utt2spk]
```

```python
df_segment_file = pd.DataFrame(utt2spk, columns =['segment_id', 'file_id'])
df_segment_file.file_id = df_segment_file.file_id.str.strip()
df_segment_file.head()
```

```python
wavscp = []
with open('./train_NPTEL_IITM/wav.scp.txt') as file:
    wavscp = file.readlines()
    
wavscp = [utt.split('\t') for utt in wavscp]
```

```python
wavscp
```

```python
df_file_wav = pd.DataFrame(wavscp, columns =['file_id', 'file_path'])
df_file_wav.file_path = df_file_wav.file_path.str.strip()
df_file_wav.head()
```

```python
df_merged = pd.merge(df_segment, df_text, on='segment_id' )

assert(len(df_merged) == len(df_segment))
```

```python


df_merged = pd.merge(df_merged, df_file_wav, on='file_id' )

assert(len(df_merged) == len(df_segment))
```

```python
df_merged['start_time'] = df_merged['start_time'].astype('float32')
df_merged['end_time'] = df_merged['end_time'].astype('float32')
```

```python
df_merged['duration'] = df_merged.end_time - df_merged.start_time
```

```python
df_merged.duration.sum() / 3600
```
